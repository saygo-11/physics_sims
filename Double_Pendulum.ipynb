# A Double Pendulum Simulator
### *An exercise in using Matplotlib animations and Lagrangian mechanics.*

Double pendulums are one of the simplest examples of chaos theory and behave according to rules that are relatively easy to encode! I will be using Lagrangian equations of motion that use energy to model motion instead of calculating each and every force.

Goals of this project:
- Learn Matplotlib animations
- Learn Lagrangian mechanics and derive equations of motion by hand
- Run Lyapunov experiment on completed simulation

Derivation:
$$\begin{align} 
L=T-V && \text{Lagrangian} 
\\ \frac{\delta}{\delta t} (\frac{\delta L}{\delta \dot{q}}) - \frac{\delta L}{\delta q} = 0 && \text{Euler-Lagrange equation}
\end{align}$$
With $q$ refering to a generalized coordinate describing the state of the system. In this case, I will use $l_1, l_2, \theta_1, \theta_2$ (the lengths of the pendulums and the angles they make with their vertical, respectively). Both $l_1, l_2$ are generally constant, but $\theta_1, \theta_2$ will change as a function of time. I begin by deriving the Kinetic Energy $T$:

$$\begin{align}
x_1 &= l_1 \sin\theta_1, \\
y_1 &= -l_1 \cos\theta_1 .
\end{align}$$

Velocities:

$$\begin{align}
\dot{x}_1 &= l_1 \dot{\theta}_1 \cos\theta_1, \\
\dot{y}_1 &= l_1 \dot{\theta}_1 \sin\theta_1.
\end{align}$$

Hence,

$$
v_1^2
= \dot{x}_1^2+\dot{y}_1^2
= l_1^2\dot{\theta}_1^2.
$$

Coordinates of the second mass:

$$\begin{align}
x_2 &= x_1 + l_2\sin\theta_2
= l_1\sin\theta_1+l_2\sin\theta_2,\\
y_2 &= y_1 - l_2\cos\theta_2
= -l_1\cos\theta_1-l_2\cos\theta_2.
\end{align}$$

Velocities:

$$\begin{align}
\dot{x}_2 &= l_1\dot{\theta}_1\cos\theta_1 + l_2\dot{\theta}_2\cos\theta_2,\\ 
\dot{y}_2 &= l_1\dot{\theta}_1\sin\theta_1 + l_2\dot{\theta}_2\sin\theta_2.
\end{align}$$

Therefore,

$$\begin{align}
v_2^2
&= \dot{x}_2^2+\dot{y}_2^2 \\
&=
l_1^2\dot{\theta}_1^2
+l_2^2\dot{\theta}_2^2
+2l_1l_2
\dot{\theta}_1\dot{\theta}_2
\left(
\cos\theta_1\cos\theta_2
+\sin\theta_1\sin\theta_2
\right) \\
&=
l_1^2\dot{\theta}_1^2
+l_2^2\dot{\theta}_2^2
+2l_1l_2
\dot{\theta}_1\dot{\theta}_2
\cos(\theta_1-\theta_2).
\end{align}$$

Kinetic Energy

$$\begin{align}
T
&=
\frac12 m_1 v_1^2
+\frac12 m_2 v_2^2 \\
&=
\frac12(m_1+m_2)
l_1^2\dot{\theta}_1^2
+m_2l_1l_2
\dot{\theta}_1\dot{\theta}_2
\cos(\theta_1-\theta_2)
+\frac12 m_2l_2^2\dot{\theta}_2^2 .
\end{align}$$

Potential Energy

Choosing the zero of potential at

$$
y=-(l_1+l_2),
$$

the heights are

$$\begin{align}
h_1 &= l_1(1-\cos\theta_1)+l_2, \\
h_2 &= l_1(1-\cos\theta_1)
+l_2(1-\cos\theta_2).
\end{align}$$

Thus,

$$\begin{align}
V
&=
m_1gh_1+m_2gh_2 \\
&=
m_1g\left[l_1(1-\cos\theta_1)+l_2\right]
+m_2g\left[
l_1(1-\cos\theta_1)
+l_2(1-\cos\theta_2)
\right].
\end{align}$$

Dropping the additive constant $m_1gl_2$ (during differentiation, it will be dropped anyways),

$$
V = (m_1+m_2)gl_1(1-\cos\theta_1) + m_2gl_2(1-\cos\theta_2).
$$


$$
\boxed{
L=T-V
}$$

or explicitly,

$$\boxed{
    \begin{aligned}
        L ={}& \frac12(m_1+m_2)l_1^2\dot{\theta}_1^2 + m_2l_1l_2 \dot{\theta}_1\dot{\theta}_2 \cos(\theta_1-\theta_2) \\
        &+ \frac12 m_2l_2^2\dot{\theta}_2^2 -(m_1+m_2)gl_1(1-\cos\theta_1) - m_2gl_2(1-\cos\theta_2).
    \end{aligned}
}$$
